# Iris Dataset Analysis
This notebook performs EDA and clustering on the iris dataset.

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
import json

# Load data
df = pd.read_csv('iris.csv')

# Descriptive Statistics
stats = df.describe().to_dict()

# Correlation matrix
corr = df.drop('variety', axis=1).corr()

# Visualizations
sns.set_theme(style="whitegrid")

# 1. Correlation Heatmap
plt.figure(figsize=(10, 8))
sns.heatmap(corr, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Iris Feature Correlation Heatmap')
plt.savefig('visualizations/correlation_heatmap.png')
plt.close()

# 2. Distributions
plt.figure(figsize=(12, 10))
for i, col in enumerate(df.columns[:-1]):
    plt.subplot(2, 2, i+1)
    sns.histplot(df, x=col, hue='variety', kde=True, element="step")
    plt.title(f'Distribution of {col}')
plt.tight_layout()
plt.savefig('visualizations/iris_distributions.png')
plt.close()

# 3. Pairplot
sns.pairplot(df, hue='variety', diag_kind='kde')
plt.savefig('visualizations/pairplot.png')
plt.close()

# 4. Clustering (K-Means)
X = df.drop('variety', axis=1)
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

kmeans = KMeans(n_clusters=3, random_state=42, n_init=10)
df['cluster'] = kmeans.fit_predict(X_scaled)

# Cluster visualization (using petal length vs petal width)
plt.figure(figsize=(10, 8))
sns.scatterplot(data=df, x='petal.length', y='petal.width', hue='cluster', style='variety', palette='viridis', s=100)
plt.title('K-Means Clustering of Iris Data (Petal Dimensions)')
plt.savefig('visualizations/iris_clusters.png')
plt.close()

# Save results for the report
results = {
    "stats": stats,
    "variety_counts": df['variety'].value_counts().to_dict(),
    "cluster_overlap": pd.crosstab(df['variety'], df['cluster']).to_dict()
}

with open('analysis_summary.json', 'w') as f:
    json.dump(results, f, indent=4)

print("Analysis complete. Visualizations and summary generated.")

```
